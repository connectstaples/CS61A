**Question**
Implement index, which takes a sequence of keys, a sequence of values, and a two-argument match function. It returns a dictionary from keys to lists in which the list for a key k contains all values v for which match(k, v) is a truthy value


````python
def index(keys, values, match):
    """Return a dict from keys k to a list of values v for which match(k, v) is a true value
    >>> index([7, 9, 11], range(30, 50), lambda k, v: v % k == 0)
    {7: [35, 42, 49], 9: [36, 45], 11: [33, 44]}
    """
    return {k: [v for v in values if match(k, v)] for k in keys}
````

#### Data Abstraction

- compound objects combine objects together 
- abstract data type lets you manipulate compound objects as units
    - it allows you to isolate two parts of any program that uses data:
        - how data are represented (as parts)
        - how data are manipulated (as units)
- Data Abstraction: methodology by which functions enforce an abstraction barrier between representation and use
    - compound data must be embraced by data abstraction to make your program more modular 
    - **Modular** def: the practice of dividing a program into separate, independent modules


**Rational Number Arithmetic Ex**
```
3/2 * 3/5 = 9/10 
3/2 + 3/5 = 21/10
```

Thoughts:
- rational(n, d) returns a rational number of x
- numer(x) returns the numerator of x 
- denom(x) returns denominator of x 


````python
"""
These functions implement an abstract data type for rational numbers
This example  mirrors the idea in SICP: abstraction allows us to separate concerns, making code more modular and maintainable
"""
from fractions import gcd 

def rational(n, d): # constructor 
    """Construct a rational number that represent N/D"""
    g = gcd(n, d)
    return [n//g, d//g]

def numer(x):
    """Return numerator of rational number X"""
    return x[0]

def denom(x):
    """Return denominator of rational number X"""
    return x[1] # select item from a list 


def mul_rational(x, y):
    return rational(numer(x) * numer(y), # rational is the CONSTRUCTOR
                    denom(x) * denom(y)) # numer, and demon, are the SELECTORS

def add_rational(x, y):
    nx, dx = numer(x), denom(y)
    ny, dy = numer(y), denom(y)
    return rational(nx * dy + ny * dx, dx * dy)

def equal_rational(x, y):
    return numer(x) * denom(y) == numer(y) * denom(x)

````

#### Tree Abstraction
![Tree](tree.png)
- By treating branches as a list, you are not violating data abstraction barrier, because you said in your abstraction, that a tree has a list of branches 


````python
"""
>>> tree(1)
[1]
>>> is_leaf(tree(1))
True
>>> t = tree(1, [tree(5, [tree(7)]), tree(6)])
>>> t
[1, [5, [7]], [6]]
>>> label(t)
1
>>> branches(t)
[[5, [7]], [6]]
>>> branches(t)[0]
[5, [7]]
>>> is_tree(branches(t)[0])
True
>>> label(branches(t)[0])
5
"""

def tree(label, branches=[]):
    for branch in branches: # verifies the tree definition
        assert is_tree(branch)
    return [label] + list(branches) # creates a list from a sequence of branches 

def label(tree):
    return tree[0]
def branches(tree):
    return tree[1:]
def is_tree(tree):
    if type(tree) != list or len(tree) < 1: # verifies that a tree is bound to a list
        return False
    for branch in branches(tree):
        if not is_tree(branch):
            return False
        return True
def is_leaf(tree):
    return not branches(tree)

def fib_tree(n):
    if n <= 1:
        return tree(n)
    else: 
        left, right = fib_tree(n-2), fib_tree(n-1)
        return tree(label(left)+label(right), [left, right])
def count_leaves(t):
    """Count leaves of tree"""
    if is_leaf(t):
        return 1
    else: return sum([count_leaves(b) for b in branches(t)])
def leaves(tree):
    """Return a list containing leaf labels of tree
    >>> leaves(fib_tree(5))
    [1, 0, 1, 0, 1, 1, 0, 1]
    """
    if is_leaf(tree):
        return [label(tree)]
    else: return sum(leaves(b)for b in branches(tree)) # list of leaf labels for each branch 

## a function that creates a tree from another tree is typically also recursive

def increment_leaves(t):
    """Return a tree like t but with leaf labels incremented"""
    if is_leaf(t):
        return tree(label(t) + 1)
    else: 
        bs = [increment_leaves(b) for b in branches(t)]
        return tree(label(t), bs)
def increment(t):
    """Return a tree like t but with leaf labels incremented"""
    return tree(label(t) + 1, [increment(b) for b in branches(t)])


def print_tree(t, indent=0):
    """This function prints the structure of the tree, instead of returning a list with sub lists"""
    print('  ' * indent + str(label(t)))
    for b in branches(t):
        print_tree(b, indent+1)

def count_paths(t, total):
    """Return number of paths from root to any node in tree t for which the laebls along the path sum to total
    >>> t = tree(3, [tree(-1), tree(1, [tree(2, tree(3)]), tree(1, [tree(-1)])])
    >>> count_paths(t, 3)
    2
    >>> count_paths(t, 4)
    2
    """
    if label(t) == total:
        found = 1
    else: 
        found = 0 
    return found + sum([count_paths(b, total - label(t)) for b in branches(t)])
````

<center> Warning Note on Lazy computation</center>

**- if you use some evaluation like == for just an object, it will return the wrong truthy value. you must assign it to a list or variable to check its boolean value** 


**Lazy Computation** 
- the program will use filter, and evaluate it so that it wont have to compute all values 
- when you call list on that, it will run the program to exhausting forcing ALL evaluations 

````python
"""
>>> def double(x):
...     print(x, '*', ' =>', 2*x)  
...     return 2*x
... 
>>> m = map(double, range(3, 7)) # map applies it to all values
>>> f = lambda y: y >= 10  # filtering 
>>> t = filter(f, m)
>>> next(t)
3 *  => 6
4 *  => 8
5 *  => 10
10
>>> next(t)
6 *  => 12
12
>>> list(t) # calling list will return no values since you already called all of them and cant retrieve them
[]

>>> list(filter(f, map(double, range(3, 7)))) # recalling but without the variable assignment since all values in m have been called on 
3 *  => 6
4 *  => 8
5 *  => 10
6 *  => 12
[10, 12]
>>> 
"""
````
